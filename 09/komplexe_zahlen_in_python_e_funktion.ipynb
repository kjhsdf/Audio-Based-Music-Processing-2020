# Komplexe Zahlen in Python


```python
a = complex(1.5, 0.5)
b = 2 + 1j
print(a, '\n')
print(b, '\n')
# addition
print(a + b, '\n')
# multiplikation
print(a * b, '\n') # (a + b) * (c + d) = a*(c + d) + b*(c + d) = ac + adi + bc + bdi
# konjugat
print(a, np.conjugate(a), '\n')
# inverse
print(a, 1 / a, a * (1 / a), '\n')
# absoluter wert
print(a, np.abs(a))
print(a, abs(a))
print(a, np.sqrt(np.real(a) ** 2 + np.imag(a) ** 2), '\n')
# winkel
print(a, np.angle(a))
print(a, np.angle(a) * 180 / np.pi)
print(a, np.arctan(np.imag(a) / np.real(a)))
print(a, np.arctan(np.imag(a) / np.real(a)) * 180 / np.pi)
c = 0 + 2j
print(c, np.arctan2(np.imag(c), np.real(c)))
print(c, np.arctan2(np.imag(c), np.real(c)) * 180 / np.pi)
```

# E-Funktion

```python
t = np.linspace(0, 7)
e = np.exp(t * 1j)

plt.plot(t, np.abs(e))
plt.plot(t, np.angle(e))
```

$e^{t \cdot 2 \pi i} = cos(t \cdot 2 \pi) + i \cdot sin(t \cdot 2 \pi)$ 

```python
t = np.linspace(0, 1, 50)
e = np.exp(t * 1j * 2 * np.pi)

plt.figure(figsize=(5 * 2, 5 * 2))
p1 = plt.subplot(2, 2, 1)
p2 = plt.subplot(2, 2, 2)
p3 = plt.subplot(2, 2, 3)
cm = plt.cm.get_cmap('viridis')
p1.set_title('e-Funktion')
p1.set_ylabel('Imag')
p1.set_xlabel('Real')
p2.set_title('Imag')
p3.set_title('Real')

for ex, tx, color in zip(e, t, cm(t)):
    p1.plot(np.real(ex), np.imag(ex), 'o', color=color)
    p2.plot(tx, np.imag(ex), 'o', color=color)
    p3.plot(np.real(ex), tx * -1, 'o', color=color)

plt.tight_layout()
plt.show()
```
