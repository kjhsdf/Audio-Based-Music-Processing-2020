```
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
import librosa as lr
```

```
x, sr = lr.load('piano_major_scale.mp3')

# normalize
x = x / np.max(np.abs(x))

display(Audio(x, rate=sr))
```

```
stft = lr.stft(x)

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
length = np.shape(x)[0]
t = np.linspace(0, length / sr, length)
plt.plot(t, x)
plt.xlim(0, length / sr)

plt.subplot(2, 1, 2)
# to dB
plt.imshow(10 * np.log10(np.abs(stft)), aspect='auto', origin='lower', extent=[0, np.shape(x)[0] / sr, 0, sr])
plt.ylim(0, sr / 2)
plt.show()
```

```
logspec = lr.pseudo_cqt(x)

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
length = np.shape(x)[0]
t = np.linspace(0, length / sr, length)
plt.plot(t, x)
plt.xlim(0, length / sr)

plt.subplot(2, 1, 2)
# to dB
plt.imshow(10 * np.log10(np.abs(logspec)), aspect='auto', origin='lower', extent=[0, np.shape(x)[0] / sr, 0, np.shape(logspec)[0]])
# plt.ylim(0, sr / 2)
plt.show()
```

```
chroma = lr.feature.chroma_stft(x)

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
length = np.shape(x)[0]
t = np.linspace(0, length / sr, length)
plt.plot(t, x)
plt.xlim(0, length / sr)

plt.subplot(2, 1, 2)
# to dB
plt.imshow(10 * np.log10(np.abs(chroma)), aspect='auto', origin='lower', extent=[0, np.shape(x)[0] / sr, 0, np.shape(chroma)[0]])
# plt.ylim(0, sr / 2)
plt.show()
```
