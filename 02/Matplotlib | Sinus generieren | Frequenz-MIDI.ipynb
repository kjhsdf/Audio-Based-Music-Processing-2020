# Frequenzen, Tonhöhen, Pitch, MIDI

```python
# 440 <=> 69 --> A' / A4

# 440 * 2 <=> 69 + 12

# 440 * 2 ** (1 / 12) <=> 69 + 1
# 440 * 2 ** (2 / 12) <=> 69 - 9

# Kommentieren: 'Cmd-/'

def midi_to_frequency(pitch):
    return 440 * 2 ** ((pitch - 69) / 12)

print(midi_to_frequency(69))
print(midi_to_frequency(81))
print(midi_to_frequency(76))
print(midi_to_frequency(60))
print(midi_to_frequency(60.5))

def frequency_to_midi(frequency):
    return np.round(12 * np.log2(frequency / 440) + 69)
#     return 12 * np.log2(frequency / 440) + 69

print()
print(frequency_to_midi(440))
print(frequency_to_midi(880))
print(frequency_to_midi(660))
```
